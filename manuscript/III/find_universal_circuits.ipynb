```python
from unicirc import search_ansatze, all_candidates_linear, filtered_candidates_linear
```

```python
num_wires = 3 # 3 qubits
num_cz = 14 # 14 entanglers (theoretical lower bound for universality on 3 qubits)
key = 21511
num_params = 9 + num_cz * 4

# Use reduction test together with simpler ``all_candidates_linear`` function
candidate_kwargs = {"reduction_test": True, "candidates_fn": all_candidates_linear} # ~16 sec
# Don't need reduction test with ``filtered_candidates_linear`` function
candidate_kwargs = {"reduction_test": False, "candidates_fn": filtered_candidates_linear} # ~8 sec

solutions = search_ansatze(
    num_wires=num_wires,
    num_cz=14,
    all_solutions=True, # Return all solutions, not just the first
    key=key,
    mat_jac_fn=None,
    rank_tol=1e-10,
    **candidate_kwargs
)
```

```python
a = (0, 1)
b = (1, 2)
max_block_sol = (a, a, b, b) * 3 + (a, a)
regular_sol = (a, b) * 7
assert max_block_sol in solutions
print("The regular solution with maximized two-qubit blocks passes the Jacobian rank test")
assert regular_sol in solutions
print("The very regular solution with alternating entanglers passes the Jacobian rank test")
```
